```python



import cv2
import numpy as np


import mediapipe as mp

mp_selfie_segmentation = mp.solutions.selfie_segmentation
selfie_segmentation = mp_selfie_segmentation.SelfieSegmentation()

# Function to create a background image of the same size as the input image, filled with the specified color.
def create_colored_background(image, color):
    background = np.zeros_like(image, dtype=np.uint8)  # pixels range dtype=np.uint8:0 to 255, an array of zeros with the same shape as the input image:zeros_like
    background[:] = color  # fills entire background image with the specified color.
    return background

# Dictionary of BGR colors
colors = {
    'white': (255, 255, 255),
    'orange': (0, 165, 255),
    'yellow': (0, 255, 255),
    'light green': (144, 238, 144),
    'pink': (203, 192, 255)
}

cap = cv2.imread('images.jpg')
if cap is None:
    raise FileNotFoundError("Input image not found. Please check the path.")

image_rgb = cv2.cvtColor(cap, cv2.COLOR_BGR2RGB)

results = selfie_segmentation.process(image_rgb)  # processes the RGB image using the selfie segmentation model, storing the result in the results variable.
mask = results.segmentation_mask  # extracts the segmentation mask from the results.

while True:
    background_choice = input("Choose background color (white, orange, yellow, light green, pink or 'q' to quit): ").strip().lower()

    if background_choice == 'q':
        break

    if background_choice not in colors:
        print("Invalid choice. Defaulting to white.")
        background_choice = 'white'

    # Create the background image based on the chosen color
    background_color = colors[background_choice]
    background_image = create_colored_background(cap, background_color)

    # Apply the mask to get the foreground
    condition = mask > 0.5
    foreground = np.where(condition[:, :, None], cap, background_image)  # to combine the original image and the background image based on the binary mask

    # Display the original and foreground images
    cv2.imshow('Original', cap)
    cv2.imshow('Foreground', foreground)

    # Use waitKey with a timeout to allow window updates
    key = cv2.waitKey(1)# returns ascii val 
    if key == ord('q'):#checks 
        break

cv2.destroyAllWindows()
```

## changing background images 

```python
import cv2
import numpy as np
import mediapipe as mp

mp_selfie_segmentation = mp.solutions.selfie_segmentation
selfie_segmentation = mp_selfie_segmentation.SelfieSegmentation()

def resize_background(background, image_shape):
    return cv2.resize(background, (image_shape[1], image_shape[0]))

# Load the image
cap = cv2.imread('images.jpg')
if cap is None:
    raise FileNotFoundError("Input image not found. Please check the path.")

background_images = {
    'background1': cv2.imread('background1.jpg'),
    'background2': cv2.imread('background2.jpg'),
    'background3': cv2.imread('background3.jpg'),
    'background4': cv2.imread('background4.jpg')
}

# Check if all background images are loaded correctly
for key in background_images:
    if background_images[key] is None:
        raise FileNotFoundError(f"Background image '{key}' not found. Please check the path.")

image_rgb = cv2.cvtColor(cap, cv2.COLOR_BGR2RGB)

results = selfie_segmentation.process(image_rgb)
mask = results.segmentation_mask

while True:
    
    background_choice = input("Choose background (background1, background2, background3, background4 or 'q' to quit): ").strip().lower()

    if background_choice == 'q':
        break

    if background_choice not in background_images:
        print("Invalid choice. Defaulting to background1.")
        background_choice = 'background1'

    # Resize the chosen background image to match the input image size
    background_image = resize_background(background_images[background_choice], cap.shape)

    condition = mask > 0.5
    foreground = np.where(condition[:, :, None], cap, background_image)

    cv2.imshow('Original', cap)
    cv2.imshow('Foreground', foreground)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
from langdetect import detect, detect_langs

# Sample texts in different languages
texts = [
    "This is a sample English text.",
    "यह एक उदाहरण हिन्दी पाठ है।",
    "ఇది ఒక నమూనా తెలుగు పాఠ్యం.",
    "இது ஒரு மாதிரி தமிழ் உரை.",
    "ഇത് ഒരു മാതൃക മലയാളം വാചകം ആണ്.",
    "これは日本語のサンプルテキストです。",
]   

# Function to detect language
def detect_language(text):
    try:
        
        lang = detect(text)
        print(f"The detected language for the text '{text}' is: {lang}")
    
        langs = detect_langs(text) #returns a list of possible languages along with their probabilities.
        print(f"Language probabilities: {langs}")
    except Exception as e:
        print(f"Error detecting language: {e}")

# Detect the language of each text
for text in texts:
    detect_language(text)
```

```python
# Define the input text that you wants the model to classify.
INPUT_TEXT = "\u5206\u4E45\u5FC5\u5408\u5408\u4E45\u5FC5\u5206" #@param {type:"string"}
```

```python
# STEP 1: Import the necessary modules.
from mediapipe.tasks import python
from mediapipe.tasks.python import text

# STEP 2: Create a LanguageDetector object.
base_options = python.BaseOptions(model_asset_path="language_detector.tflite")
options = text.LanguageDetectorOptions(base_options=base_options)
detector = text.LanguageDetector.create_from_options(options)

# STEP 3: Get the language detcetion result for the input text.
detection_result = detector.detect(INPUT_TEXT)

# STEP 4: Process the detection result and print the languages detected and
# their scores.

for detection in detection_result.detections:
  print(f'{detection.language_code}: ({detection.probability:.2f})')
```

# face mesh

```python
import cv2
import mediapipe as mp
mp_face_mesh=mp.solutions.face_mesh
mp_drawing=mp.solutions.drawing_utils
mp_drawing_styles =mp.solutions.drawing_styles

webcam=cv2.VideoCapture(0)
while webcam.isOpened():
    success,img=webcam.read()

    # face mesh model u
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    results=mp_face_mesh.FaceMesh(max_num_faces=2,refine_landmarks=True,min_detection_confidence=0.7,min_tracking_confidence=0.6).process(img) #model

   
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    if results.multi_face_landmarks: #Checks if any face landmarks were detecte
        for face_landmarks in results.multi_face_landmarks: #Iterates through the detected face landmarks.
             # drawing face Tesselation :we cover a surface with a patter
            mp_drawing.draw_landmarks(
                image=img, #img on to draw the landmark
                landmark_list=face_landmarks, #landmarks detected by the face mesh model.
                connections=mp_face_mesh.FACEMESH_TESSELATION, # tesselation connections  used to draw the triangular mesh on the face
                landmark_drawing_spec=None, #individual landmarks should be draw
                connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_tesselation_style() # predefined drawing style for the tessellation.
            )

            mp_drawing.draw_landmarks(
                image=img,
                landmark_list=face_landmarks,
                connections=mp_face_mesh.FACEMESH_CONTOURS, # facial outlines like the jawline, eyebrows, and lips
                landmark_drawing_spec=None,
                connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_contours_style()
            )

            mp_drawing.draw_landmarks(
                image=img,
                landmark_list=face_landmarks,
                connections=mp_face_mesh.FACEMESH_IRISES, # the circular outlines around the eyes.
                landmark_drawing_spec=None,
                connection_drawing_spec=mp_drawing_styles.get_default_face_mesh_iris_connections_style()
            )


    cv2.imshow("deepthi",img)
    if cv2.waitKey(20) & 0xFF == ord("q"):
        break

webcam.release()
cv2.destroyAllWindows()
```
